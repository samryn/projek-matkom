```python
!pip install cartopy
import cartopy
import cv2
import numpy as np
from skimage.feature import graycomatrix, graycoprops
import matplotlib.pyplot as plt
from google.colab import files

# Langkah 3: Unggah gambar
print("Silakan unggah gambar batik:")
uploaded = files.upload()

# Ambil nama file gambar yang diunggah
image_name = list(uploaded.keys())[0]

# Langkah 4: Muat gambar dan ubah menjadi grayscale
image = cv2.imread(image_name)
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Tampilkan gambar grayscale
plt.imshow(image_gray, cmap='gray')
plt.title("Gambar Batik dalam Grayscale")
plt.axis('off')
plt.show()

# Langkah 5: Hitung Gray Level Co-Occurrence Matrix (GLCM)
distances = [1]  # Jarak piksel
angles = [0, np.pi/4, np.pi/2, 3*np.pi/4]  # Sudut dalam radian

glcm = graycomatrix(image_gray, distances=distances, angles=angles, symmetric=True, normed=True)

# Langkah 6: Ekstraksi fitur tekstur
contrast = graycoprops(glcm, 'contrast')[0, 0]
dissimilarity = graycoprops(glcm, 'dissimilarity')[0, 0]
homogeneity = graycoprops(glcm, 'homogeneity')[0, 0]
ASM = graycoprops(glcm, 'ASM')[0, 0]
energy = graycoprops(glcm, 'energy')[0, 0]

# Tampilkan hasil fitur tekstur
print("Hasil Ekstraksi Fitur Tekstur:")
print(f"Contrast: {contrast}")
print(f"Dissimilarity: {dissimilarity}")
print(f"Homogeneity: {homogeneity}")
print(f"ASM: {ASM}")
print(f"Energy: {energy}")
```
